```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the saved model
loaded_model = load_model("deepfake_detection_model.h5")

# Function to preprocess a frame
def preprocess_frame(frame):
    # Resize the frame to match the model input size (224x224) and normalize pixel values
    frame = cv2.resize(frame, (224, 224))
    frame = frame.astype('float32') / 255.0
    return frame

# Open the video file
video_path = '/Users/nithishkumar/Documents/deepfake-video-detection-project/small dataset/deepfakes/001_870.mp4'
cap = cv2.VideoCapture(video_path)

deepfake_confidence = 0.0
original_confidence = 0.0
frame_count = 0

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Preprocess the frame
    processed_frame = preprocess_frame(frame)

    # Make predictions using the loaded model
    predictions = loaded_model.predict(np.expand_dims(processed_frame, axis=0))

    # Accumulate confidence scores
    deepfake_confidence += predictions[0][0]
    original_confidence += 1 - predictions[0][0]
    frame_count += 1

    # Display the frame with the prediction label
    cv2.putText(frame, f"Deepfake Confidence: {predictions[0][0]:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow("Video", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# Calculate average confidences
deepfake_average_confidence = deepfake_confidence / frame_count
original_average_confidence = original_confidence / frame_count

# Output the final result
threshold = 0.5
if deepfake_average_confidence > threshold:
    print("The video is classified as Deepfake.")
else:
    print("The video is classified as Original.")
```
